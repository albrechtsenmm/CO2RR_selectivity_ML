```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import os
```

```python
path_to_data = os.path.join(os.path.expanduser('~'), "Documents/CatTheory/CO2RR_selectivity_ML/Data/Datasets/BSc")

df = pd.read_csv(path_to_data + '/data', index_col=False)

standard_df = df.iloc[:, 4:-7].values

standard_df = StandardScaler().fit_transform(standard_df)

pca = PCA(n_components=4)
principalComponents = pca.fit_transform(standard_df)
```

```python
# Convert to dataframe
component_names = [f"PC{i+1}" for i in range(principalComponents.shape[1])]
X_pca = pd.DataFrame(principalComponents, columns=component_names)
```

```python
loadings = pd.DataFrame(
    pca.components_.T,  # transpose the matrix of loadings
    columns=component_names,  # so the columns are the principal components
    index=df.iloc[:, 4:-7].columns,  # and the rows are the original features
)
loadings
```

```python
nums = np.arange(1,5)
var_sum = []
var_ind = []
for num in nums:
  pca = PCA(n_components=num)
  pca.fit(standard_df)
  var_sum.append(np.sum(pca.explained_variance_ratio_))
  var_ind.append(pca.explained_variance_ratio_[-1])




plt.figure(figsize=(4,4))
plt.grid()
plt.plot(nums,var_sum,marker='o')
plt.plot(nums,var_ind,marker='o')
plt.xlabel('Principal component')
plt.ylabel('Explained variance')
plt.title('Variance explained by principal components')
```
