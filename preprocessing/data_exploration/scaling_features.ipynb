```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import os
```

```python
path_to_data = os.path.join(os.path.expanduser('~'), "Documents/CatTheory/CO2RR_selectivity_ML/Data/Datasets/BSc")

df = pd.read_csv(path_to_data + '/data', index_col=False)
```

```python
metal_df = df.iloc[0:10,:]
facet_df = df.iloc[10:,:-7]


x_facet, y_facet = facet_df["C*"].values.reshape(-1, 1), facet_df["CO*"].values.reshape(-1, 1)
x_metal, y_metal = metal_df["C*"].values.reshape(-1, 1), metal_df["CO*"].values.reshape(-1, 1)
x, y = df["C*"].values.reshape(-1, 1), df["CO*"].values.reshape(-1, 1)

# for i in range(len(df["OCCOH*"])):
#     plt.scatter(df["OCCOH*"][i], df["CO*"][i])
plt.scatter(metal_df["C*"], metal_df["CO*"], s=70, facecolors='#91c49c', edgecolors='black')
plt.scatter(facet_df["C*"], facet_df["CO*"], s=70, facecolors='#cf8985', edgecolors='black')

# linear_regressor = LinearRegression()  # create object for the class
# linear_regressor.fit(x_facet, y_facet)  # perform linear regression
# Y_pred_facet = linear_regressor.predict(x_facet)  # make predictions

# linear_regressor = LinearRegression()  # create object for the class
# linear_regressor.fit(x_metal, y_metal)  # perform linear regression
# Y_pred_metal = linear_regressor.predict(x_metal)  # make predictions

linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(x, y)  # perform linear regression
Y_pred = linear_regressor.predict(x)  # make predictions

# Adding title and labels to the axes
# plt.plot(x_facet, Y_pred_facet, color='#7f7f7f', linestyle='--')
# plt.plot(x_metal, Y_pred_metal, color='#7f7f7f', linestyle='--')
plt.plot(x, Y_pred, color='#7f7f7f')
plt.xlabel('$\Delta E_{C*} [eV]$', fontsize=14)
plt.ylabel('$\Delta E_{CO*} [eV]$', fontsize=14)
#plt.legend([metal for metal in facet_df['metal']])
plt.ylim((-2,0.49))
plt.xlim((-1,3.5))
plt.gca().spines[['right', 'top']].set_visible(False) 
plt.grid(color='black', linestyle='--')
plt.plot([-2,5],[-0.44,-0.44],'k',lw=2)
plt.text(2.1, 0.25, "$R^2$ = {:.3f}".format(r2_score(y, Y_pred)), bbox=dict(boxstyle="square",
                   ec=(1, 1, 1),
                   fc=(1, 1, 1),
                   ),
                   fontsize=12)

path_to_figs = os.path.join(os.path.expanduser('~'), "Dropbox/Apps/Overleaf/bsc_project/graphics/")

#plt.savefig(path_to_figs + 'scaling_CvCO.pdf')

plt.show()
```
