```python
import pandas as pd
import numpy as np
import os
```

```python
path_to_data = os.path.join(os.path.expanduser('~'), "Documents/CatTheory/CO2RR_selectivity_ML/Data/Datasets/BSc")

df = pd.read_csv(path_to_data + '/data', index_col=False)
```

```python
df.describe()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 

vif_df = df.iloc[:,4:-7]

# VIF dataframe 
vif_data = pd.DataFrame() 
vif_data["feature"] = vif_df.columns 
  
# calculating VIF for each feature 
vif_data["VIF"] = [variance_inflation_factor(vif_df.values, i) 
                          for i in range(len(vif_df.columns))] 
  
print(vif_data)
```

```python
features = ["Vnhe", "H*", "CO*", "OH*", "O*", "C*"]

for i in features:
    mean = df[i].mean()
    std = df[i].std()
    cv = std / mean
    print(f'The coefficient of variation for {i} is {cv}')
```
