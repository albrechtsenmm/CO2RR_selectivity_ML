```python
import matplotlib as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
```

```python
path_to_data = os.path.join(os.path.expanduser('~'), "Documents/CatTheory/CO2RR_selectivity_ML/Data/Datasets/BSc")

df = pd.read_csv(path_to_data + '/data', index_col=False)
```

```python
df.rename(columns = {'Vnhe':'V_RHE'}, inplace = True) 

corr =  df.iloc[:, 4:-7].corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

# cmap = sns.diverging_palette(250, 1, s=60, l=40, n=5, center="light", as_cmap=True)
# cmap = mpl.colormaps["coolwarm"] # see more here https://matplotlib.org/stable/users/explain/colors/colormaps.html

from matplotlib.colors import LinearSegmentedColormap, ListedColormap

cmap = LinearSegmentedColormap.from_list("",[[153/256,0/256,0/256], [1,1,1], [0/256,136/256,53/256]])

plt.figure(figsize=(10, 6))
sns.heatmap(corr, mask=mask, center=0, annot=True,
            fmt='.2f', square=True, cmap=cmap)

# path_to_figs = "../../../../Dropbox/Apps/Overleaf/BSc project/graphics/"
path_to_figs = os.path.join(os.path.expanduser('~'), "Dropbox/Apps/Overleaf/bsc_project/graphics/")

#plt.savefig(path_to_figs + 'correlation_matrix.pdf')
plt.show()
```
