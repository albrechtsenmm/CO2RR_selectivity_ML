```python
path_to_data = os.path.join(os.path.expanduser('~'), "Documents/CatTheory/CO2RR_selectivity_ML/Data/Datasets/BSc")

df = pd.read_csv(path_to_data + '/data', index_col=False)

metal_facet =  [str(df.iloc[id, 1]) for id in range(len(df))]

gene_dict = {'Cu(p111)': "CO+", 'Ag(p111)': "CO", 'Au(p111)': "CO", 'Pt(p111)': "H2", 'Pd(p111)': "H2", 'Sn(p111)': "HCOOH", 'In(p111)': "HCOOH", 'Pb(p111)': "HCOOH", 'Cd(p111)': "HCOOH", 'Fe(p111)': "H2", 'Zn(p111)': "CO", 'Cu(100)': "CO+", 'Cu(1111)': "CO+", 'Cu(911)': "CO+", 'Cu(711)': "CO+", 'Cu(511)': "CO+", 'Cu(311)': "CO+", 'Cu(111)': "CO+", 'Cu(810)': "CO+", 'Cu(610)': "CO+", 'Cu(510)': "CO+", 'Cu(310)': "CO+", 'Cu(210)': "CO+", 'Cu(755)': "CO+", 'Cu(533)': "CO+", 'Cu(211)': "CO+", 'Cu(332)': "CO+", 'Cu(331)': "CO+", 'Cu(110)': "CO+", 'Cu(650)': "CO+", 'Cu(320)': "CO+", 'Cu(210)': "CO+"}
df["Gene"] = df.apply(lambda row: gene_dict[row.metal_facet], axis=1)
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame
# Group by 'Genes' and get the size of each group
grouped = df.groupby('Gene').size()

# Find the size of the largest group
max_size = grouped.max()

# Function to oversample each group
def oversample(group):
    return group.sample(max_size, replace=True)

# Apply the function to each group and concatenate the results
oversampled_df = df.groupby('Gene').apply(oversample).reset_index(drop=True)
```

```python
numerical_df = df.iloc[:,4:]

# Find the size of the largest group
max_size = numerical_df.groupby('Gene').size().max()

# Function to oversample and add noise to each group
def oversample_with_noise(group, max_size):
    # Oversample the group
    oversampled_group = group.sample(max_size, replace=True)
    
    # Add Gaussian noise
    for col in oversampled_group.columns.drop('Gene'):
        # Mean and standard deviation for the column
        mean, std = group[col].mean(), group[col].std()
        
        # Adding Gaussian noise
        noise = np.random.normal(mean, std, size=max_size)
        oversampled_group[col] = oversampled_group[col] + noise

    return oversampled_group

# Apply the function to each group and concatenate the results
oversampled_noisy_df = numerical_df.groupby('Gene').apply(lambda x: oversample_with_noise(x, max_size)).reset_index(drop=True)
```
